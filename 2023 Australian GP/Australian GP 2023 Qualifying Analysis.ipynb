```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Australia', 'Q')
session.load()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
rus_lap = session.laps.pick_driver('RUS').pick_fastest()
ham_lap = session.laps.pick_driver('HAM').pick_fastest()

ver_lap = ver_lap.get_car_data().add_distance()
rus_lap = rus_lap.get_car_data().add_distance()
ham_lap = ham_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("Top 3 In The 2023 Australian GP Qualifying", y=0.9)

ax[0].plot(ver_lap['Distance'], ver_lap['Speed'], color='navy', label='VER')
ax[0].plot(rus_lap['Distance'], rus_lap['Speed'], color='#757575', label='RUS')
ax[0].plot(ham_lap['Distance'], ham_lap['Speed'], color='#28AE89', label='HAM')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(ver_lap['Distance'], ver_lap['Throttle'], color='navy', label='VER')
ax[1].plot(rus_lap['Distance'], rus_lap['Throttle'], color='#757575', label='RUS')
ax[1].plot(ham_lap['Distance'], ham_lap['Throttle'], color='#28AE89', label='HAM')
ax[1].set(ylabel='Throttle')

ax[2].plot(ver_lap['Distance'], ver_lap['Brake'], color='navy', label='VER')
ax[2].plot(rus_lap['Distance'], rus_lap['Brake'], color='#757575', label='RUS')
ax[2].plot(ham_lap['Distance'], ham_lap['Brake'], color='#28AE89', label='HAM')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('2023 Australian GP Qualifying.png', dpi = 350)
```
