```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Monaco', 'Q')
session.load()
```

```python
alo_lap = session.laps.pick_driver('ALO').pick_fastest()
lec_lap = session.laps.pick_driver('LEC').pick_fastest()

alo_lap = alo_lap.get_car_data().add_distance()
lec_lap = lec_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("Monaco GP 2023 Qualifying: FA14 vs CL16", y=0.9)

ax[0].plot(alo_lap['Distance'], alo_lap['Speed'], color='green', label='ALO')
ax[0].plot(lec_lap['Distance'], lec_lap['Speed'], color='#E21B4D', label='LEC')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(alo_lap['Distance'], alo_lap['Throttle'], color='green', label='ALO')
ax[1].plot(lec_lap['Distance'], lec_lap['Throttle'], color='#E21B4D', label='LEC')
ax[1].set(ylabel='Throttle')

ax[2].plot(alo_lap['Distance'], alo_lap['Brake'], color='green', label='ALO')
ax[2].plot(lec_lap['Distance'], lec_lap['Brake'], color='#E21B4D', label='LEC')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('FA14 vs CL16 2023 Monaco GP Qualifying.png', dpi = 350)
```
