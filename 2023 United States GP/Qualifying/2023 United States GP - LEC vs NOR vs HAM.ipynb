```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Austin', 'Q')
session.load()
```

```python
lec_lap = session.laps.pick_driver('LEC').pick_fastest()
nor_lap = session.laps.pick_driver('NOR').pick_fastest()
ham_lap = session.laps.pick_driver('HAM').pick_fastest()

lec_lap = lec_lap.get_car_data().add_distance()
nor_lap = nor_lap.get_car_data().add_distance()
ham_lap = ham_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("2023 United States GP - 0.15s That Covers The Top 3", y=0.9)

ax[0].plot(ham_lap['Distance'], ham_lap['Speed'], color='#00A19B', linestyle = 'dotted', label='HAM')
ax[0].plot(nor_lap['Distance'], nor_lap['Speed'], color='#FF9900', linestyle = 'dashed', label='NOR')
ax[0].plot(lec_lap['Distance'], lec_lap['Speed'], color='#FF2400', label='LEC')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(ham_lap['Distance'], ham_lap['Throttle'], color='#00A19B', linestyle = 'dotted', label='HAM')
ax[1].plot(nor_lap['Distance'], nor_lap['Throttle'], color='#FF9900', linestyle = 'dashed', label='NOR')
ax[1].plot(lec_lap['Distance'], lec_lap['Throttle'], color='#FF2400', label='LEC')
ax[1].set(ylabel='Throttle')

ax[2].plot(ham_lap['Distance'], ham_lap['Brake'], color='#00A19B', linestyle = 'dotted', label='HAM')
ax[2].plot(nor_lap['Distance'], nor_lap['Brake'], color='#FF9900', linestyle = 'dashed', label='NOR')
ax[2].plot(lec_lap['Distance'], lec_lap['Brake'], color='#FF2400', label='LEC')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('2023 United States GP - LEC vs NOR vs HAM.png', dpi = 600)
```
