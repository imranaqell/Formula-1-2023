```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Las Vegas', 'Q')
session.load()
```

```python
lec_lap = session.laps.pick_driver('LEC').pick_fastest()
sai_lap = session.laps.pick_driver('SAI').pick_fastest()

lec_lap = lec_lap.get_car_data().add_distance()
sai_lap = sai_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("2023 Las Vegas GP Qualifying - Charles Leclerc vs Carlos Sainz", y = 0.9)

ax[0].plot(lec_lap['Distance'], lec_lap['Speed'], color = '#E21B4D', label = 'LEC')
ax[0].plot(sai_lap['Distance'], sai_lap['Speed'], color = '#FCFBF4', ls = '--', label = 'SAI')
ax[0].set(ylabel = 'Speed')
ax[0].legend(loc = "lower right")

ax[1].plot(lec_lap['Distance'], lec_lap['Throttle'], color = '#E21B4D', label = 'LEC')
ax[1].plot(sai_lap['Distance'], sai_lap['Throttle'], color = '#FCFBF4', ls = '--', label = 'SAI')
ax[1].set(ylabel = 'Throttle')

ax[2].plot(lec_lap['Distance'], lec_lap['Brake'], color = '#E21B4D', label = 'LEC')
ax[2].plot(sai_lap['Distance'], sai_lap['Brake'], color = '#FCFBF4', ls = '--', label = 'SAI')
ax[2].set(ylabel = 'Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('2023 Las Vegas GP Qualifying - Charles Leclerc vs Carlos Sainz.png', dpi = 600)
```
