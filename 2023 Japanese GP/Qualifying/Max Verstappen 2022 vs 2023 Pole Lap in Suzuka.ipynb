```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
jpn22 = ff1.get_session(2022, 'Japan', 'Q')
jpn22.load()
```

```python
jpn23 = ff1.get_session(2023, 'Japan', 'Q')
jpn23.load()
```

```python
pole_22 = jpn22.laps.pick_driver('VER').pick_fastest()
pole_23 = jpn23.laps.pick_driver('VER').pick_fastest()

pole_22 = pole_22.get_car_data().add_distance()
pole_23 = pole_23.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("Max Verstappen 2022 vs 2023 Pole Lap in Suzuka", y=0.9)

ax[0].plot(pole_22['Distance'], pole_22['Speed'], color='#90E0EF', label='2022')
ax[0].plot(pole_23['Distance'], pole_23['Speed'], color='#C8A2C8', linestyle = 'dashed', label='2023')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(pole_22['Distance'], pole_22['Throttle'], color='#90E0EF', label='2022')
ax[1].plot(pole_23['Distance'], pole_23['Throttle'], color='#C8A2C8', linestyle = 'dashed', label='2023')
ax[1].set(ylabel='Throttle')

ax[2].plot(pole_22['Distance'], pole_22['Brake'], color='#90E0EF', label='2022')
ax[2].plot(pole_23['Distance'], pole_23['Brake'], color='#C8A2C8', linestyle = 'dashed', label='2023')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('Max Verstappen 2022 vs 2023 Pole Lap in Suzuka.png', dpi = 600)
```
