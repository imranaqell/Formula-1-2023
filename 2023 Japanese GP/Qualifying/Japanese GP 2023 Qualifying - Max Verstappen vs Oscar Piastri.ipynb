```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Japan', 'Q')
session.load()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
pia_lap = session.laps.pick_driver('PIA').pick_fastest()

ver_lap = ver_lap.get_car_data().add_distance()
pia_lap = pia_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("Japanese GP 2023: Max Verstappen vs Oscar Piastri", y=0.9)

ax[0].plot(ver_lap['Distance'], ver_lap['Speed'], color='navy', label='VER')
ax[0].plot(pia_lap['Distance'], pia_lap['Speed'], color='#FFF2D4', linestyle = 'dashed', label='PIA')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(ver_lap['Distance'], ver_lap['Throttle'], color='navy', label='VER')
ax[1].plot(pia_lap['Distance'], pia_lap['Throttle'], color='#FFF2D4', linestyle = 'dashed', label='PIA')
ax[1].set(ylabel='Throttle')

ax[2].plot(ver_lap['Distance'], ver_lap['Brake'], color='navy', label='VER')
ax[2].plot(pia_lap['Distance'], pia_lap['Brake'], color='#FFF2D4', linestyle = 'dashed', label='PIA')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('Japanese GP 2023 - Max Verstappen vs Oscar Piastri.png', dpi = 600)
```
