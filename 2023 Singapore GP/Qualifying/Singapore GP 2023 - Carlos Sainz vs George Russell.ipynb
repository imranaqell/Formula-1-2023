```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Singapore', 'Q')
session.load()
```

```python
rus_lap = session.laps.pick_driver('RUS').pick_fastest()
sai_lap = session.laps.pick_driver('SAI').pick_fastest()

rus_lap = rus_lap.get_car_data().add_distance()
sai_lap = sai_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("Singapore GP 2023: Carlos Sainz vs George Russell", y=0.9)

ax[0].plot(rus_lap['Distance'], rus_lap['Speed'], color='#00A19B', linestyle = 'dashed', label='RUS')
ax[0].plot(sai_lap['Distance'], sai_lap['Speed'], color='#FF2400', label='SAI')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(rus_lap['Distance'], rus_lap['Throttle'], color='#00A19B', linestyle = 'dashed', label='RUS')
ax[1].plot(sai_lap['Distance'], sai_lap['Throttle'], color='#FF2400', label='SAI')
ax[1].set(ylabel='Throttle')

ax[2].plot(rus_lap['Distance'], rus_lap['Brake'], color='#00A19B', linestyle = 'dashed', label='RUS')
ax[2].plot(sai_lap['Distance'], sai_lap['Brake'], color='#FF2400', label='SAI')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('Singapore GP 2023 - Carlos Sainz vs George Russell.png', dpi = 600)
```
