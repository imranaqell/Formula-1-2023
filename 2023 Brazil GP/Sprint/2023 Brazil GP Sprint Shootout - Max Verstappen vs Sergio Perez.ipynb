```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Brazil', 'SQ')
session.load()
```

```python
ver_lap = session.laps.pick_driver('VER').pick_fastest()
per_lap = session.laps.pick_driver('PER').pick_fastest()

ver_lap = ver_lap.get_car_data().add_distance()
per_lap = per_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("2023 Brazil GP Sprint Shootout - Max Verstappen vs Sergio Perez", y=0.9)

ax[0].plot(ver_lap['Distance'], ver_lap['Speed'], color='navy', label='VER')
ax[0].plot(per_lap['Distance'], per_lap['Speed'], color='yellow', linestyle='dashed', label='PER')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(ver_lap['Distance'], ver_lap['Throttle'], color='navy', label='VER')
ax[1].plot(per_lap['Distance'], per_lap['Throttle'], color='yellow', linestyle='dashed', label='PER')
ax[1].set(ylabel='Throttle')

ax[2].plot(ver_lap['Distance'], ver_lap['Brake'], color='navy', label='VER')
ax[2].plot(per_lap['Distance'], per_lap['Brake'], color='yellow', linestyle='dashed', label='PER')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('2023 Brazil GP Sprint Shootout - Max Verstappen vs Sergio Perez.png', dpi = 600)
```
