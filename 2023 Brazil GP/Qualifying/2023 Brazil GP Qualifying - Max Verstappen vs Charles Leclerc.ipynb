```python
import fastf1 as ff1
from fastf1 import plotting
from matplotlib import pyplot as plt

plotting.setup_mpl()
```

```python
session = ff1.get_session(2023, 'Brazil', 'Q')
session.load()
```

```python
lec_lap = session.laps.pick_driver('LEC').pick_fastest()
ver_lap = session.laps.pick_driver('VER').pick_fastest()

lec_lap = lec_lap.get_car_data().add_distance()
ver_lap = ver_lap.get_car_data().add_distance()
```

```python
fig, ax = plt.subplots(3, figsize = (16, 9))
fig.suptitle("2023 Brazil GP Qualifying - Max Verstappen vs Charles Leclerc", y=0.9)

ax[0].plot(ver_lap['Distance'], ver_lap['Speed'], color='yellow', label='VER')
ax[0].plot(lec_lap['Distance'], lec_lap['Speed'], color='red', linestyle='dashed', label='LEC')
ax[0].set(ylabel='Speed')
ax[0].legend(loc="lower right")

ax[1].plot(ver_lap['Distance'], ver_lap['Throttle'], color='yellow', label='VER')
ax[1].plot(lec_lap['Distance'], lec_lap['Throttle'], color='red', linestyle='dashed', label='LEC')
ax[1].set(ylabel='Throttle')

ax[2].plot(ver_lap['Distance'], ver_lap['Brake'], color='yellow', label='VER')
ax[2].plot(lec_lap['Distance'], lec_lap['Brake'], color='red', linestyle='dashed', label='LEC')
ax[2].set(ylabel='Brakes')

for a in ax.flat:
    a.label_outer()
    
plt.show()
```

```python
fig.savefig('2023 Brazil GP Qualifying - Max Verstappen vs Charles Leclerc.png', dpi = 600)
```
